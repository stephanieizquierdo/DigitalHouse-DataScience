```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Data/datos_previos_clean.csv',index_col=0)
```

```python
data.columns
```

```python
data['Provincia'].value_counts()
```

```python
capital_df=data.loc[data['Provincia']=='Capital Federal']
gbanorte_df=data.loc[data['Provincia']=='Bs.As. G.B.A. Zona Norte']
costa_df=data.loc[data['Provincia']=='Buenos Aires Costa Atlántica']
gbasur_df=data.loc[data['Provincia']=='Bs.As. G.B.A. Zona Sur']
gbaoeste_df=data.loc[data['Provincia']=='Bs.As. G.B.A. Zona Oeste']
cordoba_df=data.loc[data['Provincia']=='Córdoba']
santafe_df=data.loc[data['Provincia']=='Santa Fe']
bsasint_df=data.loc[data['Provincia']=='Buenos Aires Interior']
```

## Distribuciones de Precio/m2

```python
gbanorte_mean = round(gbanorte_df['price_usd_per_m2'].mean(),2)
gbasur_mean = round(gbasur_df['price_usd_per_m2'].mean(),2)
gbaoeste_mean = round(gbaoeste_df['price_usd_per_m2'].mean(),2)
capital_mean = round(capital_df['price_usd_per_m2'].mean(),2)

plt.figure(figsize=(10,10))
sns.distplot(gbanorte_df[['price_usd_per_m2']], hist=False, rug=False,label="GBA Norte ($\mu=%.2f$"+str(gbanorte_mean)+")",kde_kws = {'linewidth': 3},color='r')
sns.distplot(gbasur_df[['price_usd_per_m2']], hist=False, rug=False,label="GBA Sur ($\mu=%.2f$"+str(gbasur_mean)+")",kde_kws = {'linewidth': 3},color='g')
sns.distplot(gbaoeste_df[['price_usd_per_m2']], hist=False, rug=False,label="GBA Oeste ($\mu=%.2f$"+str(gbaoeste_mean)+")",kde_kws = {'linewidth': 3},color='b')
sns.distplot(capital_df[['price_usd_per_m2']], hist=False, rug=False,label="Capital ($\mu=%.2f$"+str(capital_mean)+")",kde_kws = {'linewidth': 3},color='orange')

plt.legend(loc='best',fontsize=16)
plt.grid()
plt.title('Distribución Capital y GBA',fontsize=20)
plt.xlabel('Precio ($/m2)',fontsize=14);

#Guardo la figura
plt.savefig('../images/Hist Capital y GBA.png')
```

```python
costa_mean = round(costa_df['price_usd_per_m2'].mean(),2)
bsasint_mean = round(bsasint_df['price_usd_per_m2'].mean(),2)
cordoba_mean = round(cordoba_df['price_usd_per_m2'].mean(),2)
santafe_mean = round(santafe_df['price_usd_per_m2'].mean(),2)

plt.figure(figsize=(10,10))
sns.distplot(costa_df[['price_usd_per_m2']], hist=False, rug=False,label="Costa ($\mu=%.2f$"+str(costa_mean)+")",kde_kws = {'linewidth': 3},color='r')
sns.distplot(bsasint_df[['price_usd_per_m2']], hist=False, rug=False,label="BsAs Int ($\mu=%.2f$"+str(bsasint_mean)+")",kde_kws = {'linewidth': 3},color='g')
sns.distplot(cordoba_df[['price_usd_per_m2']], hist=False, rug=False,label="Córdoba ($\mu=%.2f$"+str(cordoba_mean)+")",kde_kws = {'linewidth': 3},color='b')
sns.distplot(santafe_df[['price_usd_per_m2']], hist=False, rug=False,label="Santa Fe ($\mu=%.2f$"+str(santafe_mean)+")",kde_kws = {'linewidth': 3},color='orange')

plt.legend(loc='best',fontsize=16)
plt.grid()
plt.title('Distribución Combo',fontsize=20)
plt.xlabel('Precio ($/m2)',fontsize=14);

plt.savefig('../images/Hist Combo.png')
```

```python
gbanorte_mean = gbanorte_df['price_usd_per_m2'].mean()
gbasur_mean = gbasur_df['price_usd_per_m2'].mean()
gbaoeste_mean = gbaoeste_df['price_usd_per_m2'].mean()
capital_mean = capital_df['price_usd_per_m2'].mean()
costa_mean = costa_df['price_usd_per_m2'].mean()
cordoba_mean = cordoba_df['price_usd_per_m2'].mean()
santafe_mean = santafe_df['price_usd_per_m2'].mean()
bsasint_mean = bsasint_df['price_usd_per_m2'].mean()

mean_lst = ((capital_mean,gbanorte_mean,gbasur_mean,gbaoeste_mean),(cordoba_mean,santafe_mean,bsasint_mean,costa_mean))

fig, axes = plt.subplots(2, 4,figsize=(14,14))
df_lst = ((capital_df,gbanorte_df,gbasur_df,gbaoeste_df),(cordoba_df,santafe_df,bsasint_df,costa_df))
place_lst = (('Capital','GBA Norte','GBA Sur','GBA Oeste'),('Cordoba','Santa Fe','BsAs Int','Costa Atlantica'))

for i in range(2):
    for j in range(4):
        sns.distplot(df_lst[i][j][['price_usd_per_m2']], hist=False,label=place_lst[i][j],ax=axes[(i,j)])
        axes[(i,j)].axvline(mean_lst[i][j], linestyle='--')
```
